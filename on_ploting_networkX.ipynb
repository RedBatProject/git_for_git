```python
from graphviz import Graph
import random as r
```

```python
g = Graph()
```

```python
for i in range(1000):
    g.node(f'{i}')
```

```python
for i in range(1000):
    g.edge(f'{int(r.random()*1000)}',f'{int(r.random()*1000)}')
```

```python
import networkx as n
import matplotlib.pyplot as plt
```

```python
G = n.Graph()
```

```python
edges = [[int(r.random()*100),int(r.random()*100)] for i in range(100)]
```

```python
G.add_edges_from(edges)
```

```python
plt.figure(3,figsize=(12,12))
n.draw_networkx_edges(G,node_size=14,alpha=0.1,pos=n.spring_layout(G),)
plt.show()
```

```python
n.draw_spring(G)
```
